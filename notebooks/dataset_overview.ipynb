```python
import os
import librosa

folder = "../data/raw"   
files = os.listdir(folder)
durations = []

for f in files:
    if f.endswith((".mp3", ".wav", ".flac")):
        audio_path = os.path.join(folder, f)
        y, sr = librosa.load(audio_path)
        durations.append(librosa.get_duration(y=y, sr=sr))

durations
```
